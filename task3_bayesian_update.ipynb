```python
coin_head_probs= [0.12, 0.27, 0.21, 0.96] #probabilities from the task
coin_tail_probs= [1-i for i in coin_head_probs]
SIZE = len(coin_head_probs)
coin_pick_probs = [1/SIZE for _ in range(SIZE)] #in the start all coins have equal probability of being picked
tests = "H H H T H T H H H H".split(" ") #sample from the task
```

```python
head_prob = 0
for coin in range(SIZE):
     head_prob+= coin_head_probs[coin]/SIZE # formula to calculate probability of getting head on first test
```

```python
def calculate_probs(result:str):
    #The bayesian update function, take result of a test and output
    #new values for probabilities of getting head on next test, and probabilities for picked coin
    new_coin_pick_probs = []
    result_prob = head_prob if result == "H" else 1-head_prob  #expects that the input is either H for head and anything else is tails
    coin_results_probs = coin_head_probs if result == "H" else coin_tail_probs

    for coin in range(SIZE):
        new_coin_pick_prob = (coin_pick_probs[coin]*coin_results_probs[coin])/result_prob #bayesian formula
        new_coin_pick_probs.append(new_coin_pick_prob)
        
    new_head_prob = 0
    for coin in range(SIZE):
        new_head_prob += coin_head_probs[coin]*new_coin_pick_probs[coin] #general formula of probability

    return new_coin_pick_probs, new_head_prob
```

```python
test_head_probs = [round(head_prob, 2)]
for test in tests: #calculate the probabilities for all n tests (including probabilites for 1 try and n+1th)
    coin_pick_probs, head_prob = calculate_probs(test)
    test_head_probs.append(round(head_prob, 2))


print(test_head_probs)
```
